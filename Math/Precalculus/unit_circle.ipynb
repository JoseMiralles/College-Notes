## Step 1 - Getting the co-terminal angle

**PROBLEM**: Find the co-terminal angle of $- \frac{7 \pi}{6}$

In this case, we are dealing with a negative angle, which means that we have to rotate it towards the positive side until $|\theta| \leq 2\pi$.

To do this, we need to add full rotations Ex: $\theta + 2\pi$.
- If this were a positive angle, we would subtract instead.

### The Math

$$ -\frac{7\pi}{6} + 2\pi = - \frac{7\pi}{6} + \frac{2\pi}{1} $$

Find the lowest common denominator:

$$- \frac{7\pi}{6} + \frac{2\pi \cdot 6}{1 \cdot 6}
= \frac{7\pi}{6} + \frac{2\pi \cdot 6}{1 \cdot 6}
= \frac{7\pi}{6} + \frac{12\pi}{6} $$

Lastly, perform the addition:

$$ \frac{-7\pi}{6} + \frac{12\pi}{6} = \bm{\frac{5\pi}{6}} $$
