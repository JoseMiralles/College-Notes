# Functions

A function is a relationship that maps <u>1 input</u> to <u>1 output</u>.

- Input: $x$ values
    - $x$ is the independent variable.
- Output: $y$ or $f(x)$ values
    - This is the dependent variable.

## Domain and Range

Consider this set of inputs and outputs

| Work Hours | Pay   |
|:-----------|------:|
| 25         | $500  |
| 53         | $310  |
| 30         | $490  |
| 40         | $490  |

The domain would be a set of the inputs, and the range would be a set the outputs.

## Is it a Function?

This example is a function because no inputs are repeated:

`S = { (-2, 3), (-4, 3), (5, 6) }`

This example is **not** a function because inputs do repeat:

`S = { (-2, 3), (-4, 3), (-2, 6) }`


## $y^2$ leads to non-function

This is because you have to square both sides.

$y^2 = 4$

leads to:

$y = \pm \sqrt{4}$

This leads to two outputs for a single input.
